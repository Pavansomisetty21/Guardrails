```python
from guardrails_grhub_competitor_check import CompetitorCheck
import guardrails as gd
```

```python
# Generate competitors list
competitors_list = [
    "Apple",
    "Samsung",
    "Microsoft",
    "Google",
    "Alphabet",
    "Amazon",
    "Tesla",
    "Nvidia",
    "Intel",
    "Sony",
    "Huawei",
    "Dell",
    "HP",
    "Lenovo",
    "IBM",
    "Adobe",
    "Oracle"
]

# Define some text to test the validator
text = """\
In the rapidly evolving world of technology, a few giants have come to define innovation and progress,\
dominating the market with their products and services. Apple, the undisputed leader in the consumer tech space,\
has revolutionized personal devices with its iPhone, iPad, and Mac computers. Samsung, a global tech powerhouse,\
competes fiercely with Apple, offering a wide range of smartphones, tablets, and TVs that push the boundaries of display technology. \
Microsoft, once primarily known for its Windows operating system, has shifted its focus toward cloud services, gaming, and productivity tools,\
with products like Azure, Xbox, and Office 365 shaping the modern workplace. Google, through its parent company Alphabet,\
has become synonymous with search, advertising, and cloud services, while also making strides in AI, hardware, and autonomous vehicles. \
Amazon, the e-commerce titan, has diversified its reach into cloud computing with Amazon Web Services (AWS), as well as AI and streaming services.\
Tesla, under the leadership of Elon Musk, has disrupted the automotive industry with its electric vehicles, while also leading the charge in renewable energy. \
Nvidia, the leading provider of GPUs for gaming and AI applications, competes against Intel and AMD in the semiconductor market, \
while Sony and Huawei continue to innovate in consumer electronics and telecommunications. Together, these companies drive the tech industry's transformation,\
reshaping the way we live, work, and interact with technology on a global scale.\
"""
```

```python
# Create the Guard with the CompetitorCheck Validator
guard = gd.Guard().use(
    CompetitorCheck(competitors=competitors_list, on_fail="fix")
)

# Test with a given text
output = guard.parse(
    llm_output=text,
    metadata={},
)

print(output)
```
